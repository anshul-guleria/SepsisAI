## STEP 4 — Clean Clinical Text

MIMIC notes contain de-identification tokens like:

[ ** NAME ** ]<br>
[ ** DATE ** ]

```python
import re

def clean_text(text):
    
    if pd.isna(text):
        return ""
    
    text = re.sub(r"\[\*\*.*?\*\*\]", "", text)
    text = text.lower()
    
    return text
```

```python
cohort["clean_text"] = cohort["TEXT"].apply(clean_text)
```

## Step - 5 Create BERT Encoder Function

We encode text into 768-dim embeddings.

```python
# f=lambda x: len(x)
# cohort['clean_text'].apply(f)
```

```python
def encode_note(text):

    inputs = tokenizer(
        text,
        padding="max_length",
        truncation=True,
        max_length=512,
        return_tensors="pt"
    )

    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        outputs = bert_model(**inputs)

    embedding = outputs.last_hidden_state[:,0,:]

    return embedding.cpu().numpy()
```

```python
embeddings = []

for text in tqdm(cohort["clean_text"]):

    emb = encode_note(text)

    embeddings.append(emb[0])
```

```python
text_embeddings = np.array(embeddings)

print(text_embeddings.shape)
```

## Initialize Model

Transformer+cross attention

```python
import torch
import torch.nn as nn

class SepsisTransformerMultimodal(nn.Module):
    def __init__(self, lab_features=5, time_steps=24, bert_dim=768, d_model=64, nhead=4):
        super(SepsisTransformerMultimodal, self).__init__()
        
        # --- 1. Lab Time-Series Branch (Transformer) ---
        # Projects 5 lab features to d_model (64)
        self.lab_projection = nn.Linear(lab_features, d_model)
        
        # Transformer Encoder Layer
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=d_model, 
            nhead=nhead, 
            dim_feedforward=128, 
            batch_first=True
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=2)
        
        # --- 2. ClinicalBERT Branch ---
        self.bert_projection = nn.Linear(bert_dim, d_model)
        self.relu = nn.ReLU()
        
        # --- 3. Fusion & Classification ---
        self.fusion = nn.Linear(d_model + d_model, 32)
        self.classifier = nn.Linear(32, 1)

    def forward(self, x_time, x_text):
        # A. Process Labs
        # x_time shape: (Batch, 24, 5) -> (Batch, 24, 64)
        x_time = self.lab_projection(x_time)
        # Transformer output: (Batch, 24, 64)
        time_out = self.transformer_encoder(x_time)
        # Global Average Pooling over the 24 hours
        time_features = torch.mean(time_out, dim=1) 
        
        # B. Process BERT
        # x_text shape: (Batch, 768) -> (Batch, 64)
        text_features = self.relu(self.bert_projection(x_text))
        
        # C. Concatenate and Predict
        combined = torch.cat((time_features, text_features), dim=1)
        x = self.relu(self.fusion(combined))
        return self.classifier(x)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = SepsisTransformerMultimodal().to(device)
```

```python
pos_weight = torch.tensor([class_counts[0] / class_counts[1]]) # Result is ~13.6

criterion = nn.BCEWithLogitsLoss(pos_weight=pos_weight).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)
```

```python
# --- Training Loop ---
epochs = 50
for epoch in range(epochs):
    model.train()
    total_loss = 0
    
    for batch_text, batch_time, batch_labels in loader:
        # 1. Move to GPU & Match Types
        batch_text = batch_text.to(device).float()
        batch_time = batch_time.to(device).float()
        batch_labels = batch_labels.to(device).float().unsqueeze(1)

        # 2. Forward Pass
        optimizer.zero_grad()
        outputs = model(batch_time, batch_text)
        loss = criterion(outputs, batch_labels)

        # 3. Backward Pass
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    print(f"Epoch {epoch+1}/{epochs} | Avg Loss: {total_loss/len(loader):.4f}")
```

```python
import torch
import torch.nn.functional as F
from sklearn.metrics import roc_auc_score

# 1. Ensure model is in evaluation mode
model.eval()

# 2. Move data to the correct device (GPU or CPU)
# Assuming 'device' was defined during training
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

with torch.no_grad():
    # Convert numpy arrays to Tensors and move to GPU
    x_text = torch.tensor(X_text_test).float().to(device)
    x_time = torch.tensor(X_time_test).float().to(device)
    
    # Forward Pass
    logits = model(x_time, x_text)
    
    # 3. Apply Sigmoid to get probabilities (0.0 to 1.0)
    # This is required because roc_auc_score needs probabilities, not raw logits
    probs = torch.sigmoid(logits)
    
    # 4. Move back to CPU for sklearn
    y_scores = probs.cpu().numpy()
    y_true = y_test # Assuming y_test is already a numpy array or tensor

    auc = roc_auc_score(y_true, y_scores)

print(f"AUROC: {auc:.4f}")
```

```python
import torch
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report

# 1. Ensure model is in eval mode
model.eval()

with torch.no_grad():
    # 2. Prepare test data (moving to GPU)
    x_text_t = torch.tensor(X_text_test).float().to(device)
    x_time_t = torch.tensor(X_time_test).float().to(device)
    
    # 3. Get raw model outputs
    logits = model(x_time_t, x_text_t)
    
    # 4. Convert to probabilities and move to CPU
    probs = torch.sigmoid(logits).cpu().numpy()

# 5. Apply the 0.35 threshold
pred_binary = (probs > 0.35).astype(int)

# 6. Generate the matrix
# Ensure y_test is a 1D numpy array
y_true = y_test.numpy() if torch.is_tensor(y_test) else y_test
cm = confusion_matrix(y_true, pred_binary)

print("Confusion Matrix:")
print(cm)

# 7. Bonus: Print the report for context
print("\nClassification Report:")
print(classification_report(y_true, pred_binary))
```
